```python
pd.set_option('display.max_rows', None)
def data_overview(b):
    overview = pd.DataFrame({
        "Column": b.columns,
        "Data Type": b.dtypes.astype(str).values,
        "Total Rows": len(b),
        "Missing": b.isnull().sum().values,
        "% Missing": (b.isnull().mean().values * 100).round(2),
        "Unique Values": b.nunique().values,
        "Unique Sample": [
            [x.item() if hasattr(x, "item") else x for x in b[col].dropna().unique()[:5]]
            for col in b.columns]
    })
    
    overview = overview.reset_index(drop=True)
    return overview

data_overview(b)
```

```python
# Fitur 1: pernah nunggak atau tidak (flag)
b['FLAG_EVER_OVERDUE'] = b['AMT_CREDIT_MAX_OVERDUE'].notna().astype(int)

# Fitur 2: kalau pernah, berapa besarnya — sisanya isi 0
b['AMT_CREDIT_MAX_OVERDUE'] = b['AMT_CREDIT_MAX_OVERDUE'].fillna(0)
```

```python
data_overview(b)
```
