# SQL Querry Join and Aggregation From Home Credit Dataset using BigQuery

SELECT * FROM `HomeCredit.installments_payments_cleaned` limit 10;

SELECT * FROM `HomeCredit.application_train_cleaned` limit 10;


CREATE OR REPLACE TABLE `HomeCredit.application_train_cleaned_joined` AS
WITH agg_installments_payments_cleaned AS (
    SELECT
        SK_ID_CURR,
        AVG(DAYS_ENTRY_PAYMENT - DAYS_INSTALMENT) AS AVG_PAYMENT_LATENCY,
        MAX(DAYS_ENTRY_PAYMENT - DAYS_INSTALMENT) AS MAX_PAYMENT_LATENCY,
        SUM(AMT_INSTALMENT - AMT_PAYMENT) AS TOTAL_AMT_DIFF,
        AVG(AMT_INSTALMENT - AMT_PAYMENT) AS AVG_AMT_DIFF
    FROM `HomeCredit.installments_payments_cleaned`
    GROUP BY SK_ID_CURR
)
SELECT
    app.*,
    ins.AVG_PAYMENT_LATENCY,
    ins.MAX_PAYMENT_LATENCY,
    ins.TOTAL_AMT_DIFF,
    ins.AVG_AMT_DIFF
FROM `HomeCredit.application_train_cleaned` app
LEFT JOIN agg_installments_payments_cleaned ins ON app.SK_ID_CURR = ins.SK_ID_CURR;

SELECT * FROM `HomeCredit.application_train_cleaned_joined` limit 10;





CREATE OR REPLACE TABLE `HomeCredit.application_train_cleaned_joined` AS
WITH agg_pos_cash_balance_cleaned AS (
    SELECT
        SK_ID_CURR,
        AVG(SK_DPD) AS AVG_DPD,
        AVG(SK_DPD_DEF) AS AVG_DPD_DEF,
        AVG(CNT_INSTALMENT_FUTURE) AS AVG_INSTALMENT_FUTURE,
        MAX(NAME_CONTRACT_STATUS) AS LAST_CONTRACT_STATUS
    FROM `HomeCredit.POS_CASH_balance_cleaned`
    GROUP BY SK_ID_CURR
)
SELECT
    app.*,
    pos.AVG_DPD,
    pos.AVG_DPD_DEF,
    pos.AVG_INSTALMENT_FUTURE,
    pos.LAST_CONTRACT_STATUS
FROM `HomeCredit.application_train_cleaned_joined` app
LEFT JOIN agg_pos_cash_balance_cleaned pos ON app.SK_ID_CURR = pos.SK_ID_CURR;

SELECT * FROM `HomeCredit.application_train_cleaned_joined` limit 10;



CREATE OR REPLACE TABLE `HomeCredit.application_train_cleaned_joined` AS

WITH agg_credit_card_balance_cleaned AS (
    SELECT
        SK_ID_CURR,
        AVG(SK_DPD)                                            AS CC_AVG_DPD,
        AVG(AMT_BALANCE / NULLIF(AMT_CREDIT_LIMIT_ACTUAL, 0)) AS CC_AVG_UTILIZATION,
        AVG(AMT_PAYMENT_CURRENT)                               AS CC_AVG_PAYMENT,
        AVG(AMT_INST_MIN_REGULARITY)                           AS CC_AVG_MIN_PAYMENT,
        AVG(CNT_DRAWINGS_ATM_CURRENT)                          AS CC_AVG_ATM_DRAWINGS
    FROM `HomeCredit.credit_card_balance_cleaned`
    GROUP BY SK_ID_CURR
)

SELECT
    app.*,
    cc.CC_AVG_DPD,
    cc.CC_AVG_UTILIZATION,
    cc.CC_AVG_PAYMENT,
    cc.CC_AVG_MIN_PAYMENT,
    cc.CC_AVG_ATM_DRAWINGS

FROM `HomeCredit.application_train_cleaned_joined` app
LEFT JOIN agg_credit_card_balance_cleaned cc ON app.SK_ID_CURR = cc.SK_ID_CURR;

SELECT * FROM `HomeCredit.application_train_cleaned_joined` limit 10;




CREATE OR REPLACE TABLE `HomeCredit.application_train_cleaned_joined` AS
WITH agg_bureau_cleaned AS (
    SELECT
        SK_ID_CURR,
        MAX(CREDIT_DAY_OVERDUE) AS MAX_CREDIT_DAY_OVERDUE,
        AVG(CREDIT_DAY_OVERDUE) AS AVG_CREDIT_DAY_OVERDUE,
        MAX(AMT_CREDIT_MAX_OVERDUE) AS MAX_AMT_CREDIT_MAX_OVERDUE,
        SUM(AMT_CREDIT_SUM_OVERDUE) AS TOTAL_AMT_CREDIT_SUM_OVERDUE,
        SUM(CNT_CREDIT_PROLONG) AS TOTAL_CNT_CREDIT_PROLONG
    FROM `HomeCredit.bureau_cleaned`
    GROUP BY SK_ID_CURR
)
SELECT
    app.*,
    bur.MAX_CREDIT_DAY_OVERDUE,
    bur.AVG_CREDIT_DAY_OVERDUE,
    bur.MAX_AMT_CREDIT_MAX_OVERDUE,
    bur.TOTAL_AMT_CREDIT_SUM_OVERDUE,
    bur.TOTAL_CNT_CREDIT_PROLONG
FROM `HomeCredit.application_train_cleaned_joined` app
LEFT JOIN agg_bureau_cleaned bur ON app.SK_ID_CURR = bur.SK_ID_CURR;

SELECT * FROM `HomeCredit.application_train_cleaned_joined` limit 10;






CREATE OR REPLACE TABLE `HomeCredit.application_train_cleaned_joined` AS
WITH agg_previous_application_cleaned AS (
    SELECT
        SK_ID_CURR,
        MAX(NAME_CONTRACT_STATUS) AS LAST_CONTRACT_STATUS_PREV,
        MAX(CODE_REJECT_REASON) AS LAST_REJECT_REASON,
        AVG(DAYS_DECISION) AS AVG_DAYS_DECISION
    FROM `HomeCredit.previous_application_cleaned`
    GROUP BY SK_ID_CURR
)
SELECT
    app.*,
    prev.LAST_CONTRACT_STATUS_PREV,
    prev.LAST_REJECT_REASON,
    prev.AVG_DAYS_DECISION
FROM `HomeCredit.application_train_cleaned_joined` app
LEFT JOIN agg_previous_application_cleaned prev ON app.SK_ID_CURR = prev.SK_ID_CURR;
